```python
#import labaries to graph the df 

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
 
```

```python
# SQLAlchemy
from sqlalchemy import create_engine
import pandas as pd

#apply password 

pg_key = os.getenv('PGPASSWORD')

# Path to sqlite
database_path = 'postgresql://postgres:postgres@localhost:5432/employees_data_base'

```

# Establishing a Connection with SQL DB using SQLAlchemy

```python
# Create an engine that can talk to the database
engine = create_engine(f"{database_path}")
```

```python
# Create a connection to the engine

connection = engine.connect()
```

# Importing Tables from SQL and Displaying Them 

```python
# run a query using the data base

# import employees table

employees = pd.read_sql('SELECT * FROM employees', connection, parse_dates=['birth_date','hire_date'])
employees.head(5)



```

```python
# Import Salaries Table

salaries = pd.read_sql('SELECT * FROM salaries', connection, parse_dates=['from_date','to_date'])
salaries.head(5)
```

```python
#importing titles table
titles = pd.read_sql('SELECT * FROM titles', connection, parse_dates=['from_date','to_date'])
titles.to_date = titles['to_date'].fillna(pd.to_datetime('2030-01-01'))

titles.head(5)
```

```python
#importing department_employee table

dept_emp = pd.read_sql('SELECT * FROM dept_emp', connection, parse_dates = ['from_date','to_date'])
dept_emp.to_date = dept_emp['to_date'].fillna(pd.to_datetime('2030-01-01'))

dept_emp.head()
```

```python
#importing department table

departments = pd.read_sql('SELECT * FROM departments', connection)

departments.head()
```

# Merging Table to Create Most Common Salary Ranges 


```python
#Merging tables 

salaries_title_df = titles.merge(salaries,on = 'emp_no')

salaries_title_df.head()
```

```python
salarysummary_df = salaries_title_df[["emp_no","title","salary"]]

salarysummary_df.head()
```

```python
# Elaborate Histogram showing the numbers of empoyees according to salaries

fig_salaries = plt.hist(salarysummary_df['salary'])
fig_salaries = plt.xlabel('Range of Salaries')
fig_salaries = plt.ylabel('Number of Employees')
plt.savefig('Salaries_Distribution.png')
# plt.show()
```

```python
salaries_title_df.head()
```

# Merging Table to Create Dept Salary Graph



```python
dept_salaries_df = dept_emp.merge(salaries, on = 'emp_no')
dept_salaries_df.head()



```

```python
department_df = dept_salaries_df.merge(departments, on = 'dept_no')
department_df.head()
```

```python
#creating clean data from for avg salary per department

dep_salary_df = department_df[['salary','dept_name']]
dep_salary_df.head()

# salarysummary_df = salaries_title_df[["emp_no","title","salary"]]

# salarysummary_df.head()
```

```python
# Group by and generating the average salary by department 

dep_salary_df.groupby('dept_name').mean().round(decimals=0)

avg_dep_salary = dep_salary_df.groupby('dept_name').mean().round(decimals=0)

avg_dep_salary.sort_values(by ='salary', ascending=False)
```

```python
dep_salary_df.groupby('dept_name').mean().round(decimals=0).plot(kind='bar', color='green')
plt.savefig('Avg Salary by Dept.png')
plt.show()
```

# Create a bar chart of average salary by title.

```python
salary_title = salaries_title_df[['title','salary']]

avgsalary_title = salary_title.groupby('title').mean().round(decimals=0)
avgsalary_title.head()
avgsalary_title.sort_values(by='salary', ascending=False)

```

```python
avgsalary_title.groupby('title').mean().round(decimals=0).plot(kind='bar', color='blue')
plt.savefig('Avg Salary by Dept.png')
plt.show()
```
